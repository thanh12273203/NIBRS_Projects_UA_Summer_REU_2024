```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python

data = {
    'Nashville': {'Volume Ratio': 0.51, 'Correlation': 0.67},
    'Boston':  {'Volume Ratio': 0.02, 'Correlation': -0.03},
    'Chicago':  {'Volume Ratio': 0.32, 'Correlation': 0.57}
}
data = pd.DataFrame.from_dict(data, orient='index')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(5, 3))

sns.heatmap(
    data=data,
    vmin=-1,
    vmax=1.0,
    cmap='coolwarm',
    linewidths=0.1,
    linecolor='k',
    annot=True,
    fmt='.2f',
    cbar=True,
)
ax.tick_params(
    axis='y',
    labelrotation=0,
)

plt.savefig(
    'heatmap.png',
    bbox_inches='tight',
)
plt.show()
```
